## File Name : "month_sales.csv"

### Demo Output :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
monthly_sales = pd.read_csv("month_sales.csv")
```

```python
monthly_sales
```

```python
quart_1 = monthly_sales[:3]
quart_2 = monthly_sales[2:6]
quart_3 = monthly_sales[5:9]
quart_4 = monthly_sales[8:12]
```

```python
fig,(ax1 , ax2 , ax3 , ax4) = plt.subplots(4,1, figsize = (7,9))
ax1.plot(monthly_sales["Months"] , monthly_sales["sales"], color = "#ffa089", linewidth = 2.8)
ax2.plot(monthly_sales["Months"] , monthly_sales["sales"], color = "#ffa089", linewidth = 2.8)
ax3.plot(monthly_sales["Months"] , monthly_sales["sales"], color = "#ffa089", linewidth = 2.8)
ax4.plot(monthly_sales["Months"] , monthly_sales["sales"], color = "#ffa089", linewidth = 2.8)

ax1.set_yticklabels([])
ax1.set_xticklabels([])
         
ax2.set_xticklabels([])
ax2.set_yticklabels([])

ax3.set_yticklabels([])
ax3.set_xticklabels([])

ax4.set_yticklabels([])
ax4.set_xticklabels([])

ax1.tick_params(bottom = 0, left = 0)
ax2.tick_params(bottom = 0, left = 0)
ax3.tick_params(bottom = 0, left = 0)
ax4.tick_params(bottom = 0, left = 0)

angles = ['right' , "left" , "top", "bottom"]
         
for i in angles:
    ax1.spines[i].set_visible(False)
    ax2.spines[i].set_visible(False)
    ax3.spines[i].set_visible(False)
    ax4.spines[i].set_visible(False)

          
ax1.plot(quart_1["Months"] , quart_1["sales"] , color = "#0C3CA6" , linewidth = 2.8)
ax2.plot(quart_2["Months"] , quart_2["sales"] , color = "#0C3CA6" , linewidth = 2.8)
ax3.plot(quart_3["Months"] , quart_3["sales"] , color = "#0C3CA6" , linewidth = 2.8)
ax4.plot(quart_4["Months"] , quart_4["sales"] , color = "#0C3CA6" , linewidth = 2.8)
         
# Jan - March
ax1.text(0.5, 20000,"14236.9", alpha = 0.7, color ="#0C3CA6") 
ax1.text(2.5 , 60000 , "55691.0" , alpha = 0.7, color ="#0C3CA6") 
# April - June
ax2.text(2.5 , 60000,"55691.0", alpha = 0.7, color ="#0C3CA6") 
ax2.text(5.5, 40000, "34595.1" , alpha = 0.7, color ="#0C3CA6") 
# July - Sept
ax3.text(5.9 , 40000 , "34595.1" , alpha = 0.7, color ="#0C3CA6")
ax3.text(8.8 , 90000 , "81777.3" , alpha = 0.7, color ="#0C3CA6")
# Oct - December
ax4.text(8.8 , 90000 , "81777.3" , alpha = 0.7, color ="#0C3CA6")
ax4.text(11.8 , 60000 , "69545.6" , alpha = 0.7, color ="#0C3CA6")

ax1.text( 2.3, 800, "Jan-Mar" , rotation = 70 , color ="#0C3CA6")
ax2.text( 3.5 , 9000 , "Apr-Jun" , rotation = -10 ,color ="#0C3CA6")
ax3.text( 7.5 , 50000 , "Jul-Sep" , rotation = 60 ,color ="#0C3CA6")
ax4.text( 10 , 60000 , "Oct-Dec" , rotation = 60 , color ="#0C3CA6")

ax1.text( 0.5 , 100000 , "Quarterly Performance Review of Market" ,\
          fontdict = {'family' : 'serif' ,'size' : 17 , 'weight' : 'bold'})
ax1.text( 0.5 , 90000 , "Total Sales in each Quater" ,\
          fontdict = {'size' : 10})

monthly_sales['Cumulative_Sales'] = monthly_sales['sales'].cumsum()
sum_of_quart1 = round(monthly_sales['sales'][0:3].sum(),2)
sum_of_quart2 = round(monthly_sales['sales'][3:6].sum(),2)
sum_of_quart3 = round(monthly_sales['sales'][6:9].sum(),2)
sum_of_quart4 = round(monthly_sales['sales'][9:12].sum(),2)

ax1.text(5,1900, "Total sales in Q1 = " + str(sum_of_quart1) ,\
          fontdict = { "color" : "#0C3CA6", "size" : 9,})
ax2.text(5,1900, "Total sales in Q2 = " + str(sum_of_quart2) ,\
          fontdict = { "color" : "#0C3CA6", "size" : 9,})
ax3.text(5,1900, "Total sales in Q3 = " + str(sum_of_quart3) ,\
          fontdict = { "color" : "#0C3CA6", "size" : 9,})
ax4.text(5,1900, "Total sales in Q4 = " + str(sum_of_quart3) ,\
          fontdict = { "color" : "#0C3CA6", "size" : 9,})
plt.show()
```
